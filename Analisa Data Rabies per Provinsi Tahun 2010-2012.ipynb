# 1. Deskripsi Data

Data ini berisi mengenai jumlah kasus rabies beserta jenisnya per provinsi dari tahun 2010-2012. Data bersumber dari Ditjen PP & PL, Kemenkes RI tahun 2013.

Penjelasan mengenai Variabel pada Data ini:
1. kode_provinsi       : Kode Provinsi
2. nama_provinsi       : Nama Provinsi
3. tahun               : Tahun terjadi kasus rabies
4. jenis_rabies        : Jenis Rabies
5. jumlah_kasus_rabies : Jumlah Kasus Rabies dari jenis tersebut

Keterangan yang terdapat pada data ini adalah:
1. GHPR = Gigitan Hewan Penular Rabies (belum confirmed lab)
2. VAR = Kasus digigit yang diberi Vaksin Anti Rabies
3. LYSSA = Positif rabies dan mati

```python
# Pemanggilan data
import pandas as pd
data_rabies = pd.read_csv('situasi-rabies-per-provinsi-2010-2012.csv', delimiter = ',', encoding = "ISO-8859-1")
data_rabies
```

# 2. Statistik Tentang Data

```python
# Mencari Jumlah kasus rabies yang terjadi dari tahun 2010-2012

total_kasus=data_rabies.jumlah_kasus_rabies.sum()

print("Kasus rabies yang terjadi dari tahun 2010-2012 sebanyak", total_kasus ,"kasus")
```

```python
# Menampilkan data kasus rabies yang dikelompokkan berdasarkan Kode Provinsi, Nama Provinsi, Tahun dan Jenis Rabies

data_rabies.groupby(['kode_provinsi','nama_provinsi','tahun','jenis_rabies','jumlah_kasus_rabies']).mean()
```

```python
# Menampilkan data kasus rabies yang dikelompokkan berdasarkan Kode Provinsi, Nama Provinsi, dan Tahun

data_rabies.groupby(['kode_provinsi','nama_provinsi','tahun']).sum() ['jumlah_kasus_rabies']
```

```python
# Menampilkan data rata-rata kasus rabies dalam setiap tahun

data_rabies.groupby('tahun').mean() ['jumlah_kasus_rabies']
```

```python
# Menampilkan data jumlah kasus rabies dalam setiap tahun

data_rabies.groupby('tahun').sum() ['jumlah_kasus_rabies']
```

```python
# Menampilkan data jumlah kasus rabies berdasarkan jenis rabies

data_rabies.groupby('jenis_rabies').sum() ['jumlah_kasus_rabies']
```

```python
# Menampilkan jumlah data kasus rabies yang dikelompokkan berdasarkan Nama Provinsi

data_rabies.groupby(['kode_provinsi','nama_provinsi']).sum() ['jumlah_kasus_rabies']
```

# 3. Visualisasi Sederhana

```python
# 1. Diagram batang kasus rabies berdasarkan tahun
ax = data_rabies.groupby('tahun').mean()['jumlah_kasus_rabies'] #ditampung ke object ax
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tahun", figsize=(6,8), legend=True, fontsize=12) #kind=jenis diagram, title= judul
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12) 
```

```python
# 2. Diagram batang kasus rabies berdasarkan jenis rabies

ax = data_rabies.groupby('jenis_rabies').mean()['jumlah_kasus_rabies'] #ditampung ke object ax
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Jenis", figsize=(6,8), legend=True, fontsize=12) #kind=jenis diagram, title= judul
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
# 3. Diagram batang kasus rabies berdasarkan provinsi

ax = data_rabies.groupby('nama_provinsi').mean()['jumlah_kasus_rabies'] #ditampung ke object ax
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Provinsi", figsize=(15,2), legend=True, fontsize=12) #kind=jenis diagram, title= judul
ax.set_xlabel("Nama Provinsi", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```
